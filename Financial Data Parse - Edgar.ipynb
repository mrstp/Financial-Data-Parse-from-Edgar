```python
import pandas as pd
import requests
```

```python
url = "https://www.sec.gov/Archives/edgar/data/1326801/000132680122000108/meta-20220930.htm"

request = requests.get(url, headers={'user-agent': 'Mozilla/5.0'})

df = pd.read_html(request.text)
```

```python
len(df)
```

# **Income Statement**

```python
for table in df:
    IS = (table[0].str.contains("Earnings per share attributable to Class A and Class B common stockholders:"))
    if IS.any():
        Income_Statement = table

Income_Statement = Income_Statement.iloc[1:,[2,4,10,16,22]]
Income_Statement
```

```python
header = Income_Statement.iloc[0] + " " + Income_Statement.iloc[1]
Income_Statement.columns = header
```

```python
Income_Statement = Income_Statement.iloc[2:]
```

```python
Income_Statement[Income_Statement.columns[1:]] = Income_Statement[Income_Statement.columns[1:]].astype(str)
```

```python
cols = [1,2,3,4]
for i in cols:
    Income_Statement[Income_Statement.columns[i]] = Income_Statement[Income_Statement.columns[i]].map(lambda x: x.replace("(","-"))
    Income_Statement[Income_Statement.columns[i]] = Income_Statement[Income_Statement.columns[i]].map(lambda x: x.replace(")",""))
```

```python
Income_Statement[Income_Statement.columns[1:]] = Income_Statement[Income_Statement.columns[1:]].astype(float)
Income_Statement
```

# **Balance Sheet**

```python
for table in df:
    BS = (table[0].str.contains("Liabilities and stockholders' equity"))
    if BS.any():
        Balance_Sheet = table

Balance_Sheet = Balance_Sheet.iloc[1:,[2,4,10]]
Balance_Sheet
```

```python
Balance_Sheet.columns = Balance_Sheet.iloc[0]
Balance_Sheet[Balance_Sheet.columns[1:]] = Balance_Sheet[Balance_Sheet.columns[1:]].astype(str)
```

```python
cols = [1,2]
for i in cols:
    Balance_Sheet[Balance_Sheet.columns[i]] = Balance_Sheet[Balance_Sheet.columns[i]].map(lambda x: x.replace("(","-"))
    Balance_Sheet[Balance_Sheet.columns[i]] = Balance_Sheet[Balance_Sheet.columns[i]].map(lambda x: x.replace(")",""))
    Balance_Sheet[Balance_Sheet.columns[i]] = Balance_Sheet[Balance_Sheet.columns[i]].map(lambda x: x.replace("—","0"))
    Balance_Sheet[Balance_Sheet.columns[i]] = Balance_Sheet[Balance_Sheet.columns[i]].map(lambda x: x.replace(",",""))
        
Balance_Sheet = Balance_Sheet.iloc[1:]
Balance_Sheet[Balance_Sheet.columns[1:]] = Balance_Sheet[Balance_Sheet.columns[1:]].astype(float)
Balance_Sheet
```
